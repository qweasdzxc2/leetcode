459 重复的子字符串
===
给定一个非空的字符串 s ，检查是否可以通过由它的一个子串重复多次构成

题解1 双循环暴力解
---
- 如果s能够由它的字串重复拼接获得，那么s的长度length必然是字串的倍数
- 在1到length/2之间查找符合是length因数的数i
- 从i开始，在i之前的串中查找对应位置是否相等
- 出现不对应就直接结束循环并返回false

题解2 移动匹配
---
- 拼接原字符串 s 形成 s+s
- 对拼接字符串去掉第一个字符和最后一个字符，不去除会出现使用s直接匹配到构成拼接的两个字符串前s或后s，因为显然s+s由s组成，s必然是其中的一部分
- 使用库函数匹配查找去掉第一个字符和最后一个字符的 s+s 中是否存在s

925 长按键入
===
你的朋友正在使用键盘输入他的名字 name。偶尔，在键入字符 c 时，按键可能会被长按，而字符可能被输入 1 次或多次。

你将会检查键盘输入的字符 typed。如果它对应的可能是你的朋友的名字（其中一些字符可能被长按），那么就返回 True。

题解
---
- typed中的字符有且只有两种用途
    - 正常键入
    - 长按重复键入
- 判断typed中的字符是否和name对应顺序的字符相同
- 不相同则判断是否和typed的前一个字符相同
- 是则说明是长按重复
- 否则说明键入有误，直接返回false
- 遍历完typed时，若name的指针数值上等于name的长度，说明name的所有字符都按顺序包含在typed中，返回true

844 比较含退格的字符串
===
给定 s 和 t 两个字符串，当它们分别被输入到空白的文本编辑器后，如果两者相等，返回 true 。# 代表退格字符。

注意：如果对空文本输入退格字符，文本继续为空。

题解1 栈模拟
---
- 利用栈后进先出的特性
- 使用ArrayList集合模拟栈
    - 字符串遍历到 # 且集合元素不为0，就把集合最后一个元素删除
    - 集合转数组返回
- 使用库函数Arrays.equals()比较两个字符数组是否相等，可以先行判定两个数组是否为空，都为空必然相等

题解2 从后往前双指针
---
- 从字符串最后开始遍历，记录遍历遇到的退格字符 #
- 遇到一个 # 就让指针多前移一位
- 处理完退格后，比较两个字符串各自的指针位置的字符是否相等
- 没有同时遍历完或对应位置不相同时，返回false
