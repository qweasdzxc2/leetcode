257 二叉树的所有路径
===
给你一个二叉树的根节点 root ，按 任意顺序 ，返回所有从根节点到叶子节点的路径。

叶子节点 是指没有子节点的节点。

题解1 深度优先遍历
---
- 使用递归的前序遍历
- 是叶子结点时，则将当前结点和之前的path拼接，作为String加入List集合
- 每一次递归使用的都是新的path，因为String是不可变的
- 非叶子节点则继续递归遍历，分别遍历左右孩子
- 采用非递归的前序遍历则考虑栈即可，整体逻辑不变
- 广度优先同理，将遍历过程转为广度优先即可，整体逻辑不变

题解2 回溯算法
---
- 回溯算法的本质是穷举，部分情况可根据需要进行剪枝操作
- 本题采用回溯+前序遍历的解法
- 回溯函数没有返回值，本题结束条件是当前节点为空
- 若当前节点不为空，则将当前长度记录length，因为有可能当前节点是叶子节点，之后需要回溯到这个位置，之后将当前节点的值记录到path
- 若当前节点是叶子节点，则将path加入List集合
- 若当前节点不是，则先在path加入一个 -> ，之后考虑左右孩子，即前序遍历的顺序，直到好到叶子节点
- 找到叶子节点之后就可以考虑回溯，退回到上一个位置

100 相同的树
===
给你两棵二叉树的根节点 p 和 q ，编写一个函数来检验这两棵树是否相同。

如果两个树在结构上相同，并且节点具有相同的值，则认为它们是相同的。

题解
---
- 对称树的变体，考虑递归，同时遍历
- 先分别考虑四种情况：
    - 当前节点都为空 --- true
    - 左边的树当前节点为空，右边的树当前节点非空 --- false
    - 左边的树当前节点非空，右边的树当前节点为空 --- false
    - 两棵树的当前节点值不同 --- false
- 递归遍历比较两棵树的对应结点，返回比较值即可
