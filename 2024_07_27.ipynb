18 四数之和
===
给你一个由 n 个整数组成的数组 nums ，和一个目标值 target 。请你找出并返回满足下述全部条件且不重复的四元组 [nums[a], nums[b], nums[c], nums[d]] （若两个四元组元素一一对应，则认为两个四元组重复）：

0 <= a, b, c, d < n
a、b、c 和 d 互不相同
nums[a] + nums[b] + nums[c] + nums[d] == target
你可以按 任意顺序 返回答案 。

题解
---
- 类似三数之和的解法，先对数组进行排序
- 特殊条件判断数组长度不足4位的直接返回空集合
- 第一层循环获得第一个数字，遍历终止在倒数第四个数，为了确保后三个数能正常通过下标索引获取，需要去重
- 当前数字和它之和的三个数字之和大于target时，无需继续查找元组，因为继续后移得到的元组必然大于这四数之和
- 当前数字和数组最后三个数字之和小于target时，说明需要更大的第一个数字
- 第二层循环获取第二个数字，遍历终止在倒数第三个数，确保后两个数能正常取值，需要去重
- 当第一个数和第二个数及它之和两个数之和大于target，无需继续查找元组
- 当第一个数和第二个数与最后两个数之和小于target，说明需要更大的第二个数
- 双指针法获取第三个和第四个数，第四个数无需考虑去重，因为前三个数都判断了去重

290 单词规律
===
给定一种规律 pattern 和一个字符串 s ，判断 s 是否遵循相同的规律。

这里的 遵循 指完全匹配，例如， pattern 里的每个字母和字符串 s 中的每个非空单词之间存在着双向连接的对应规律。

题解
---
- 问题转化为判断pattern和s是否存在双射关系
- 使用两个map集合记录，集合c2s键值对以规律pattern为key，字符串s为value；集合s2c键值对以字符串s为key，pattern为value
- 同时遍历，判断双射是否成立


205 同构字符串
===
给定两个字符串 s 和 t ，判断它们是否是同构的。

如果 s 中的字符可以按某种映射关系替换得到 t ，那么这两个字符串是同构的。

每个出现的字符都应当映射到另一个字符，同时不改变字符的顺序。不同字符不能映射到同一个字符上，相同字符只能映射到同一个字符上，字符可以映射到自己本身。

题解
---
- 两个map集合判断是否满足双射
- 解法同290

1002 查找常用字符
===
给你一个字符串数组 words ，请你找出所有在 words 的每个字符串中都出现的共用字符（ 包括重复字符），并以数组形式返回。你可以按 任意顺序 返回答案。

题解
---
- 由于题目限制字符串出现的字符均为小写字母，使用两个长度为26的数组存储出现次数
- 遍历字符串数组words，获得其中的每一个字符串
- 将字符串的字母出现次数存储在数组freq中，并根据数组freq更新结果数组minFreq
- 取对应位置的较小值更新minFreq，其中minFreq初始化存入最大值
- 遍历完所有字符串后，遍历结果数组，根据结果数组对应位置存的元素存入输出结果的集合中，注意个数
