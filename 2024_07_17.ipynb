242 有效的字母异位词
===
给定两个字符串 s 和 t ，编写一个函数来判断 t 是否是 s 的字母异位词。

注意：若 s 和 t 中每个字符出现的次数都相同，则称 s 和 t 互为字母异位词。

题解1 哈希表法(适用于所有字符)
---
- 当s和t长度不同时，直接返回false
- 使用map集合存储，字符作为key，字符出现的次数作为value
- 先遍历字符串s，记录在map集合
- 使用getOrDefault() 方法，该方法能够获取指定 key 对应对 value，如果找不到 key ，则返回设置的默认值
- 遍历字符串t，使用t出现的字符作为key查找对应的value并减一
- 如果出现value小于0说明t出现了s未出现过的字符或出现的次数不相同，返回false
- 字符串t遍历完成后未出现value小于0的情况，说明s和t互为字母异位词，返回true

题解2 字典法
---
- 创建并维护一个长度为26的数组，存放对应的字符出现次数
- 遍历字符串s更新数组，对应位置元素+1
- 遍历字符串t将对应位置的元素-1
- 遍历数组，若出现元素小于0，返回false
- 遍历完成后返回true

349 两个数组的交集
===
给定两个数组 nums1 和 nums2 ，返回 它们的 交集 。

输出结果中的每个元素一定是 唯一 的。我们可以 不考虑输出结果的顺序 。

题解
---
- 两个set集合，setNum用于对nums1去重，setArr用于对交集结果去重
- 遍历数组nums1，存入setNum可保证无重复元素
- 遍历数组nums2，在setNum中查找是否出现过，出现则存入setArray
- 将setArray转为数组输出

202 快乐数
===
编写一个算法来判断一个数 n 是不是快乐数。

「快乐数」 定义为：

对于一个正整数，每一次将该数替换为它每个位置上的数字的平方和。
然后重复这个过程直到这个数变为 1，也可能是 无限循环 但始终变不到 1。
如果这个过程 结果为 1，那么这个数就是快乐数。
如果 n 是 快乐数 就返回 true ；不是，则返回 false 。

题解1 哈希表去重法
---
- 由于限制了n的取值范围，当n的位数得到13位的时候，循环的最大值都不会超过1053，因此每次用于平方和计算的数位数大于2小于13的情况下，平方和的计算结果都会是三位，而n的范围限制了这个数最大只有(2^31 - 1)，即只有10位
- 进行数位分离，计算平方和作为下一个用于计算的n
- 使用set集合判断是否出现重复数字，即进入循环，进入则返回false，否则继续计算直到进入循环或返回1
- 出现1则返回true
