21 合并两个有序链表
===
- 将两个升序链表合并为一个新的 升序 链表并返回。
- 新链表是通过拼接给定的两个链表的所有节点组成的。

题解1 迭代
---
- 创建的结点用于表示合并后的链表，新建另一结点指向这个新结点并用于遍历
- 在两个链表都不为零时同时遍历两个原链表，比较结点存储数据的大小
- 将结点数据较小的那一个链接到临时结点之后
- 循环直到遍历到其中一个链表的最后
- 将另一链表的剩余部分全部拼接到新链表之后
- 返回新链表的首元结点

题解2 递归
---
- 1.终止条件：其中一个链表为空
- 2.返回值：需要拼接上的链表结点
- 3.单层需要做的操作：比较两个链表结点的值的大小，并将值较小的那一个拼接到链表后面

328 奇偶链表
===
- 给定单链表的头节点 head ，将所有索引为奇数的节点和索引为偶数的节点分别组合在一起，然后返回重新排序的列表。
- 第一个节点的索引被认为是 奇数 ， 第二个节点的索引为 偶数 ，以此类推。
- 请注意，偶数组和奇数组内部的相对顺序应该与输入时保持一致。
- 你必须在 O(1) 的额外空间复杂度和 O(n) 的时间复杂度下解决这个问题。

题解
---
- 声明一个偶数节点链表的头结点结点指向偶数结点的首元结点
- 奇数结点odd和奇数链表头结点head
- 偶数结点even和偶数链表头结点evenhead
- 奇数结点的下一个结点是偶数结点的下一个结点
- 偶数结点的下一个结点是奇数结点的下一个结点
- 最后将偶数链表拼接到奇数链表尾部
- 返回头结点head即表示重新排序后的链表

206 反转链表1
===
- 给你单链表的头节点 head ，请你反转链表，并返回反转后的链表。

题解1 递归
---
- 1.结束边界：head指向的结点为空或head的下一个结点为空
- 2.返回值：返回反转后的链表
- 3.单层操作：将head.next.next 赋值为 head, 此时出现环形链表，将head.next赋值为null，解除环形
- 代码随想录--翻转链表

题解2 迭代
---
- 使用三个指针分别是pre = null；node = head； next = head.next
- 当next指针指向的不为空时，执行断链反转和三指针后移
- next指针指向null时，说明node指向原链表的最后一个结点，此时执行最后一次断链反转
- 返回node即表示反转后的链表

题解3 头插法
---
- 定义一个新的链表头结点newHead
- 遍历原链表
- 保存当前结点的下一个结点用于移动
- 将当前结点插入新的链表头部
- 新的头结点变为使用头插法插入的结点，即当前遍历到的结点
- 移动到下一个结点继续遍历
- 返回新的链表头结点即可

92 反转链表2
===
- 给你单链表的头指针 head 和两个整数 left 和 right ，其中 left <= right 。
- 请你反转从位置 left 到位置 right 的链表节点，返回 反转后的链表 。

题解1 穿针引线
---
- 使用哑节点dummyHead，便于操作首元结点
- 四个指针：
1. pre指向待反转部分的前驱结点
2. leftNode指向待反转部分的首元结点
3. succ指向待反转部分的后继结点
4. rightNode指向待反转部分的尾结点
- 确定四个结点位置后，执行断链，反转
- 三部分链表拼接
- 返回dummyHead.next即可


题解2 头插法
---
- 三个指针
1. pre指向待反转部分的前驱结点
2. curr指向待反转区域的首元结点
3. next指向curr的下一个结点
