24 两两交换链表中的节点
===
- 给你一个链表，两两交换其中相邻的节点，并返回交换后链表的头节点。
- 你必须在不修改节点内部的值的情况下完成本题（即，只能进行节点交换）

题解1 递归
---
[递归套路](https://lyl0724.github.io/2020/01/25/1/)
- 1.终止条件：链表只剩一个节点或无节点
- 2.返回值：已经进行过交换处理的链表
- 3.判断本层操作需要完成的事

- 定义节点head和节点next，其中节点next是节点head指向的下一个节点
- 终止条件是链表只剩一个节点或者无节点
- 函数返回已经处理好的部分，由于是链表，返回起始节点即可
- 需要执行的操作是，把head和next的位置交换，next指向已处理好的链表部分


题解2 迭代
---
- 定义空节点指向原链表的首元节点
- 确保每次迭代时，节点剩余个数大于2
- 执行节点交换
- 每轮交换执行完成后，标志节点temp后移
- 返回新链表

160 相交链表
===
- 给你两个单链表的头节点 headA 和 headB ，请你找出并返回两个单链表相交的起始节点。
- 如果两个链表不存在相交节点，返回 null 。
- 题目数据 保证 整个链式结构中不存在环

解法1 集合哈希
---
- 利用set集合的contains()方法判断集合是否存在特定元素
- 先将链表A的节点存入set集合visited
- 遍历链表B，使用contains()方法判断B中的节点是否在visited中出现
- 若遍历结束仍未出现，则返回null

解法2 双指针遍历
---
- 定义两个指针nodeA和nodeB用于遍历链表A和链表B
- 当其中任意一个链表遍历完后，其对应的指针重新指向另一个链表的起始节点开始遍历
- 定义链表A长度为m，链表B长度为n，即使两个链表长度不同，在最坏情况下，也能在O(m+n)的时间复杂度下完成
- 若两条链表没有相交节点，则在O(m+n)的时间后返回null

203 移除链表元素
===
- 给你一个链表的头结点 head 和一个整数 val 
- 请你删除链表中所有满足 Node.val == val 的节点，并返回 新的头节点 。

解法1 递归
---
- 先对除了首元结点以外的节点判断是否等于val
- 1.结束条件，链表遍历到为空，即head == null
- 2.返回值，返回已经删除好的链表
- 3.一次操作内容：判断是否符合删除的条件，符合则将节点指向下一个节点，不符合则保留

解法2 迭代
---
- 需要定义一个新的节点dummyHead指向原链表的首元节点
- 设置用于迭代的临时节点temp
- 遍历链表，当符合删除条件时，将temp.next = temp.next.next，即直接断链跳过要删除的节点
- 不符合删除条件则继续后移temp = temp.next
- 函数返回链表的首元节点的dummyHead.next

19 删除链表的倒数第N个结点
===
- 给你一个链表，删除链表的倒数第 n 个结点，并且返回链表的头结点。

解法1 计算链表长度
---
- 哑结点的作用：便于操作链表时不需要额外考虑首元结点，哑结点dummyHead指向链表的首元节点，dummyHead.next即表示原链表
- 第一次遍历链表，计算出链表的长度
- 第二次遍历链表，遍历到要删除的结点前一个位置
- 使目标删除结点的前一个结点后继节点变为目标删除结点的后继
- 返回dummyHead.next

解法2 递归
---
- 设定全局变量num，赋初始值为0，每次递归完成后则自增
- 由于递归传值从最后往前调用，当num == n时，即为需要删除的结点，跳过即可
- 1.终止条件：链表为空
- 2.返回值：链表首元节点
- 3.不是要删除的结点则不进行任何操作

解法3 双指针
---
- 设置哑结点dummyHead便于操作原链表头结点
- 设置指针slow和fast
- fast先移动到n+1的位置
- slow和fast同时移动直到fast指向null
- 此时令slow.next = slow.next.next,即断链删除
- 返回dummyHead.next
