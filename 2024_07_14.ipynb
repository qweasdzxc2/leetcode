142 环形链表2
===
给定一个链表的头节点  head ，返回链表开始入环的第一个节点。 如果链表无环，则返回 null。

如果链表中有某个节点，可以通过连续跟踪 next 指针再次到达，则链表中存在环。 

为了表示给定链表中的环，评测系统内部使用整数 pos 来表示链表尾连接到链表中的位置（索引从 0 开始）。如果 pos 是 -1，则在该链表中没有环。注意：pos 不作为参数进行传递，仅仅是为了标识链表的实际情况。

不允许修改 链表。

题解1 哈希集合
---
- 使用哈希集合存储链表结点
- 遍历链表，判断链表的结点是否在集合中
- 如果存在直接返回这个结点，即为环形链表入环第一个结点
- 如果不存在，则结点加入集合
- 链表遍历完成后仍然没有集合元素匹配，则不是环形链表，返回null

题解2 双指针遍历
---
- 慢指针slow速度是快指针fast的二分之一
- 若无环，快指针会先出现为空的情况，跳出遍历返回null


234 回文链表
===
给你一个单链表的头节点 head ，请你判断该链表是否为回文链表

如果是，返回 true ；否则，返回 false 。

题解 链表元素存入数组遍历
---
- 将链表元素存入数组
- 使用双指针left和right分别指向数组的起始位置和结束位置
- 依次判断数组left位置和right位置的数据是否相同
- 若出现不相同则返回false
- 若成功遍历完成数组，则说明是回环，返回ture

143 重排链表
===
给定一个单链表 L 的头节点 head ，单链表 L 表示为：

L0 → L1 → … → Ln - 1 → Ln
请将其重新排列后变为：

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
不能只是单纯的改变节点内部的值，而是需要实际的进行节点交换。

题解1 链表转线性表
---
- 将链表元素存入线性表
- 分别使用两个指针left和right从起始位置和结束位置相向遍历线性表
- 将left指向的结点的next指向right指向的结点
- 遍历结束后将最后一个结点指向null

题解2 寻找链表中点 + 链表逆序 + 合并链表
---
- 找到链表中间结点
- 中间结点之后的链表成为新的链表
- 新链表反转
- 使用# 21的方法进行链表合并
- 返回合并后的链表


876 链表的中间结点
===
给你单链表的头结点 head ，请你找出并返回链表的中间结点。

如果有两个中间结点，则返回第二个中间结点。

题解1 双指针遍历
---
- 慢指针slow一次走一步，快指针fast一次走两步
- 当fast指向null时，slow必然是中间结点或两个中间结点的第二个中间结点

题解2 数组
---
- 单链表转数组
- 取数组的二分之一长度位置的元素即为所求结点
