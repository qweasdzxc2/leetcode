2 两数相加
===
- 给你两个 非空 的链表，表示两个非负的整数。它们每位数字都是按照 逆序 的方式存储的，并且每个节点只能存储 一位 数字。
- 请你将两个数相加，并以相同形式返回一个表示和的链表。
- 你可以假设除了数字 0 之外，这两个数都不会以 0 开头。

题解
---
- 创建两个结点分别为head和tail，赋初始值为null
- 同时遍历两个链表，由于是逆序存储，当两个链表长度不同是，较短的那一条看作拼接了多个0
- 结点的数值sum为，链表对应结点之和加进位值并对10取模
- 进位值carry为sum / 10，carry取值在0或1
- 在第一次相加之后，head指向tail(即新的链表的首元结点)并保持不变
- 之后相加只移动指向原链表的两个指针取值和指向结果链表的指针赋值
- 返回head即表示结果链表

394 字符串解码
===
- 给定一个经过编码的字符串，返回它解码后的字符串。
- 编码规则为: k[encoded_string]，表示其中方括号内部的 encoded_string 正好重复 k 次。注意 k 保证为正整数。
- 你可以认为输入字符串总是有效的；输入字符串中没有额外的空格，且输入的方括号总是符合格式要求的。
- 此外，你可以认为原始数据不包含数字，所有的数字只表示重复的次数 k ，例如不会出现像 3a 或 2[4] 的输入。

题解1 双栈类逆波兰表达式法
---
- 分别定义栈stackNum存储数字和stackStr存储字符，并定义字符串str存储解码后的字符串
- 遍历原字符串s，考虑以下四种情况：
1. 数字
- 考虑数字是否有多位
- 加入数字栈stackNum
2. 左括号[
- 将当前累计的字符串入栈
- 重置str
3. 右括号]
- 以数字栈的栈顶元素为循环次数，获得拼接字符
- 拼接完成之后，再拼接上字符栈stackStr的栈顶元素，赋值给str
4. 字符
- 拼接在str之后
- 返回str

题解2 单栈操作
---
- 左括号和字符正常入栈
- 数字判断具体位数后整体入栈
- 右括号不入栈，作为出栈开始条件
- 数字字符串拼接位置
