20 有效括号
===
- 给定一个只包括 '('，')'，'{'，'}'，'['，']' 的字符串 s ，判断字符串是否有效。
- 有效字符串需满足：
1. 左括号必须用相同类型的右括号闭合。
2. 左括号必须以正确的顺序闭合。
3. 每个右括号都有一个对应的相同类型的左括号。

题解
---
- 定义Map集合paris存储括号对，便于查找，key为右括号，value为对应的左括号的值
- 定义栈stack
- 遍历字符串s
- 判断当前字符的括号类型：左括号则进栈
- 是右括号，若当前栈为空或当前栈顶元素不是该括号对应的左括号，则可确定该字符串s无法匹配，返回false
- 以上条件都满足则栈顶元素出栈匹配
- 直到字符串s遍历完成
- 返回值为栈satck.isEmpty();
- 如果栈空，说明全部出栈，全部匹配完成；如果栈非空，说明有左括号剩余，匹配未完成

155 最小栈
===
- 设计一个支持 push ，pop ，top 操作，并能在常数时间内检索到最小元素的栈。
- 实现 MinStack 类:
1. MinStack() 初始化堆栈对象。
2. void push(int val) 将元素val推入堆栈。
3. void pop() 删除堆栈顶部的元素。
4. int top() 获取堆栈顶部的元素。
5. int getMin() 获取堆栈中的最小元素。

题解1 辅助栈
---
- 定义两个栈，xStack存储所有入栈的元素，minStack存储xStack每次执行入栈时栈中最小值
- minStack初始化一个栈元素类型最大值入栈，便于后续取最小值时的比较
- 执行出栈时，两个栈都进行出栈
- 取最小值时，直接将minStack栈顶元素出栈
