429 N叉树的层序遍历
===
给定一个 N 叉树，返回其节点值的层序遍历。（即从左到右，逐层遍历）。

树的序列化输入是用层序遍历，每组子节点都由 null 值分隔（参见示例）。

题解
---
- 本质是层序遍历修改
- 由于孩子结点使用List集合存储，在从队列中取出结点时，判断该结点的孩子集合是否未空即可
- 双层循环，因此边进边出不影响结果，外层循环次数是第一次计算的队列长度，因此后续加入的孩子结点不影响结果
- 内层循环根据出队列的结点的孩子结点个数确定，即List集合的大小

515 在每个树行中找最大值
===
给定一棵二叉树的根节点 root ，请找出该二叉树中每一层的最大值。

题解1 广度优先
---
- 层序遍历，初始化 max = INTEGER.MIN_VALUE
- 使用max存储每行的最大值
- 遍历完每一行将max存入结果集合即可

题解2 深度优先
---
- 递归前序遍历
- 遍历时先判断当前结点的值和list集合中的值哪个大，替换成较大的值
- 之后递归继续遍历左子树和右子树
