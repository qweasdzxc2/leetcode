根据next数组进行匹配思路
===
1. 初始化指针i和j，指针i是主串s的指针，指针j是模式串t的指针，i和j都初始化为0，从各自的串首开始移动
2. 使用指针i遍历主串s，KMP算法的特性是主串匹配的指针不回退
3. 遍历时进行条件判断：
    - 当主串字符和模式串字符匹配时，两个指针都后移
    - 当主串字符与模式串字符失配时，将模式串指针移动到 根据前一个匹配上的模式串字符的next数组值 的位置
    - 单独讨论主串第一个字符和模式串第一个字符就失配的情况，此时主串指针继续前移，模式串指针不动
4. 在遍历时判定，当模式串的最后一位与主串匹配时，即 j = s.length() 时结束匹配，返回 i-j 即为子串匹配开始位置

next[]数组生成
===
1. 初始化 next[0] = 0， 指向前缀最后一个字符的指针 j = 0
2. 从 i = 1 位置开始遍历模式串t给next[]数组赋值：
    - 当前缀与后缀相同时，j++
    - 当前缀与后缀不相同时，指针j执行回退，退到 j = next[j-1] 位置
    - 将 next[i] = j
3. 遍历t结束后获得next[]数组
