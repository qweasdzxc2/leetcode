559 N叉树的最大深度
===
给定一个 N 叉树，找到其最大深度。

最大深度是指从根节点到最远叶子节点的最长路径上的节点总数。

N 叉树输入按层序遍历序列化表示，每组子节点由空值分隔（请参见示例）。

题解
---
- 递归后序遍历
- 遍历子树，获取子树中深度的最大值
- 返回结果加一，因为当前节点是作为子树的根节点，需要加上

110 平衡二叉树
===
给定一个二叉树，判断它是否是 
平衡二叉树

题解
---
- 本质是计算二叉树的结点高度，后序遍历
- 当一个结点的左右子树高度差大于1是，使该结点的高度置为-1，此时可以说明不是平衡二叉树
- 否则返回结点的左右子树中，较大的高度+1作为结点的高度
