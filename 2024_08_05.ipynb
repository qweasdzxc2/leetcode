199 二叉树的右视图
===
给定一个二叉树的 根节点 root，想象自己站在它的右侧，按照从顶部到底部的顺序，返回从右侧所能看到的节点值。

题解
---
- 本质是层序遍历
- 使用队列作为辅助，初始化二叉树根节点进入队列
- 之后分层遍历，每次要执行出队列时，实际上队列中的元素都是同一层的
- 因此只需要获取队列的当前大小，并且获得队尾元素，非队尾元素直接出队列并判断是否有左右子节点，有则把子节点进队列
