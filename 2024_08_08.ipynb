226 反转二叉树
===
给你一棵二叉树的根节点 root ，翻转这棵二叉树，并返回其根节点。

题解1 使用深度优先遍历
---
- 前序遍历和后序遍历均可实现，中序遍历不行，因为会导致部分结点的左右孩子进行了两次反转
- 可以使用递归法遍历或迭代法遍历，

题解2 使用广度优先遍历
---
- 层序遍历实现，模板+结点交换即可

101 对称二叉树
===
给你一个二叉树的根节点 root ， 检查它是否轴对称。

题解
---
- 使用非严格的后序遍历，左子树遍历顺序：左右中，右子树遍历顺序：右左中
- 递归三部曲
    - 返回类型时boolean，返回对称轴两侧结点值是否相等，参数是左右两个子结点
    - 条件判断：
        - 左空右非空 false
        - 左非空右空 false
        - 左空右空 true
        - 左右非空但值不相等 false
    - 操作：对比左右子树的外侧是否相等，内侧是否相等

222 完全二叉树的结点个数
===
给你一棵 完全二叉树 的根节点 root ，求出该树的节点个数。

完全二叉树 的定义如下：在完全二叉树中，除了最底层节点可能没填满外，其余每层节点数都达到最大值，并且最下面一层的节点都集中在该层最左边的若干位置。若最底层为第 h 层，则该层包含 1~ 2h 个节点。

题解1 按照普通二叉树遍历解
---
- 层序遍历，与普通二叉树相同，每次出队列时更新结点计数值即可

题解2 利用完全二叉树性质解
---
- 完全二叉树：在完全二叉树中，除了最底层节点可能没填满外，其余每层节点数都达到最大值，并且最下面一层的节点都集中在该层最左边的若干位置。若最底层为第 h 层，则该层包含 1~ 2^(h-1)  个节点。
- 根据以上定义，说明在完全二叉树中，左子树的深度可以代表整棵树的深度
- 完全二叉树只有两种情况：
    - 满二叉树
    - 最后一层叶子结点没有满
        - 那么接着遍历必然能找到满二叉树，因为单独一个结点也能作为满二叉树
- 使用递归获取当前结点左右子树的高度
- 如果相等，则根据公式计算以当前节点为根节点的二叉树的结点个数
- 如果不相等，则递归直到找到满二叉树，计算结点个数
