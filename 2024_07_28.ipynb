不要给另外的数组分配额外的空间，你必须原地修改输入数组、使用 O(1) 的额外空间解决这一问题。

541 反转字符串Ⅱ
===
给定一个字符串 s 和一个整数 k，从字符串开头算起，每计数至 2k 个字符，就反转这 2k 字符中的前 k 个字符。

如果剩余字符少于 k 个，则将剩余字符全部反转。
如果剩余字符小于 2k 但大于或等于 k 个，则反转前 k 个字符，其余字符保持原样。

题解
---
- 模拟处理
- 指针从下标0开始，以每次k*2的步长移动
- 每次指针移动执行字符串部分反转
- 反转的起点为指针当前指向位置，终点为数组长度与(当前位置+k)中的较小值-1
- 由于题目给的时String，在java中时不可变的，需要转为char[]类型的数组进行反转

54 替换数字（卡码网）
===
给定一个字符串 s，它包含小写字母和数字字符，请编写一个函数，将字符串中的字母字符保持不变，而将每个数字字符替换为number。 例如，对于输入字符串 "a1b2c3"，函数应该将其转换为 "anumberbnumbercnumber"。

题解
---
- 遍历一次字符串记录需要转成number的数字字符个数
- 根据字符个数扩充字符串数组到合适的大小
- 数组初始化依次填入字符串s的字符，可调用java库函数进行复制
- 双指针遍历，new初始化指向数组最后一个位置，old初始化指向原字符串最后一个元素填充进的位置
- 从后往前遍历数组，new和old每次都移动1位
- 当old指针遍历到数字时，从后往前依次填充“r”、“e”、“b”、“m”、“u”、“n”，new移动到“n”的位置
- 从后往前遍历的原因：降低时间复杂度，预先流出足够的数组空间，无需在填充时对原字符串数字之后的元素进行移动

151 反转字符串中的单词
===
给你一个字符串 s ，请你反转字符串中 单词 的顺序。

单词 是由非空格字符组成的字符串。s 中使用至少一个空格将字符串中的 单词 分隔开。

返回 单词 顺序颠倒且 单词 之间用单个空格连接的结果字符串。

注意：输入字符串 s中可能会存在前导空格、尾随空格或者单词间的多个空格。返回的结果字符串中，单词间应当仅用单个空格分隔，且不包含任何额外的空格。

题解1 使用库函数
---
- 使用库函数trim()对字符串s进行去除首尾空格
- 将连续的空格作为分割条件，对字符串s进行split()分割，使用正则表达式匹配连续空格
- 结果转为List集合以便调用Colections的reverse方法进行集合元素反转
- 返回结果使用join在各个集合元素中间加上空格即可
